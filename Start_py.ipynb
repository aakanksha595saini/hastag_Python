```python
#Armstrong Numbers in a Range - Find all Armstrong numbers (sum of each digit raised to the power of total digits = number) between 100 and 999.
for num in range(100,999):
  sum = 0
  temp = num
  while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10
  if num == sum:
    print(num,"is an Armstrong number")
  #else:
    #print(num,"is not an Armstrong number")
```
